```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import os
```

```python
def preprocess_data(df: pd.DataFrame) -> pd.DataFrame:
    df_c = df.explode("events", ignore_index=True)
    df_c = pd.concat([df_c.drop(["events"], axis=1), df_c["events"].apply(pd.Series)], axis=1)
    return df_c
```

```python
TRAIN_TS_START = 1659726648560
TRAIN_TS_END = 1660304817956

TEST_TS_START = TRAIN_TS_END
TEST_TS_END = 1660608701141
```

```python
os.system(f"mkdir /kaggle/working/otto_exploded_dataset")

for it in ["clicks", "carts", "orders"]:
    os.system(f"mkdir /kaggle/working/otto_exploded_dataset/{it}")

for dt in ["train", "test"]:
    for it in ["clicks", "carts", "orders"]:
        os.system(f"mkdir /kaggle/working/otto_exploded_dataset/{it}/{dt}")
```

```python
chunks = pd.read_json(
    "/kaggle/input/otto-recommender-system/train.jsonl",
    lines=True,
    chunksize=2**15
)

for i, df in enumerate(tqdm(chunks)):
    df_pp = preprocess_data(df)
    for itype in ["clicks", "carts", "orders"]:
        print(itype)
        df_pp_t = df_pp[df_pp["type"] == itype]
        
        df_pp_t_train = df_pp_t[df_pp_t["ts"].between(TRAIN_TS_START, TRAIN_TS_END, "left")]
        df_pp_t_test = df_pp_t[df_pp_t["ts"].between(TEST_TS_START, TEST_TS_END, "left")]
    
        df_pp_t_train.to_parquet(
            "/kaggle/working/otto_exploded_dataset/" + itype + f"/train/part_{i}.parquet"
        )
        df_pp_t_test.to_parquet(
            "/kaggle/working/otto_exploded_dataset/" + itype + f"/test/part_{i}.parquet"
        )
```
